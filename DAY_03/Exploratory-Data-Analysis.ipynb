```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("India_Menu.csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
data.shape
```

```python
data.columns
```

```python
data.nunique()
```

```python
data['Sodium (mg)'].nunique()
```

```python
#Let's clean the data
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
India_Menu = data.drop(['Added Sugars (g)', 'Total Sugars (g)'], axis = 1)
```

```python
India_Menu.head()
```

```python
India_Menu.mean()
```

```python
India_Menu.mode()
```

```python
India_Menu.median()
```

```python
#Relationship analysis
```

```python
corelation = India_Menu.corr()
```

```python
plt.figure(figsize=(18,12))
plt.style.use('fivethirtyeight')
sns.heatmap(corelation, xticklabels = corelation.columns, yticklabels = corelation.columns, annot = True)
plt.show()
```

```python
sns.pairplot(India_Menu)
plt.show()
```

```python
sns.relplot(x = 'Energy (kCal)', y = 'Sat Fat (g)', hue = 'Protein (g)', data = India_Menu)
plt.figure(figsize=(18,8))
plt.show()
```

```python
sns.distplot(India_Menu['Cholesterols (mg)'], bins = 27)
plt.show()
```

```python
sns.catplot(x = 'Sodium (mg)' , kind = 'box', data = India_Menu)
plt.show()
```

```python
sns.violinplot(x = data['Total carbohydrate (g)'] , y = data['Total Sugars (g)'])
plt.show()
```

```python
sns.swarmplot(x = data['Protein (g)'] , y = data['Total Sugars (g)'])
plt.show()
```
