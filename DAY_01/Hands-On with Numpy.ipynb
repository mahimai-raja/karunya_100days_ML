## Arrays


A numpy array is a grid of values, all of the same type, and is indexed by a tuple of nonnegative integers. The number of dimensions is the rank of the array; the shape of an array is a tuple of integers giving the size of the array along each dimension.

We can initialize numpy arrays from nested Python lists, and access elements using square brackets:

```python
dir(np)
```

**One Dimension Array**

```python
import numpy as np

a = np.array([1,2,3,4,5,6])
print("a:",a)
print(type(a))
print("a.shape:",a.shape)
a[4]=10
print("a:",a)
print("a.size:",a.size)
print(a[4])
```

# Slicing arrays

Slicing in python means taking elements from one given index to another given index.

We pass slice instead of index like this: [start:end].

We can also define the step, like this: [start:end:step].

If we don't pass start its considered 0

If we don't pass end its considered length of array in that dimension

If we don't pass step its considered 1

```python
arr = np.array([1, 2, 3, 4, 5, 6, 7])
print(arr[:]) # arr[:]
print(arr[1:5]) 
print(arr[1:])
print(arr[2:4])
```

# numpy.sum()

The numpy.sum() function is available in the NumPy package of Python. This function is used to compute the sum of all elements, the sum of each row, and the sum of each column of a given array.

```python
import numpy as np  
a=np.array([0.4,0.5])  
b=np.sum(a)  
b  
```

```python
import numpy as np  
a=np.array([[1,4],[3,5]])  
b=np.sum(a,axis=0)  #the sum of the column elements has been calculated accordingly.
b
```

# concatenate()

The concatenate() function is a function from the NumPy package. This function essentially combines NumPy arrays together. This function is basically used for joining two or more arrays of the same shape along a specified axis. 

```python
x=np.array([[1,2],[3,4]])  
y=np.array([[12,30]])  

z=np.concatenate((x,y))  

print(z)
```

```python
x=np.array([[1,2],[3,4]])  
y=np.array([[12,30]])

print(x)

#Transpose
print(y.T)  

z=np.concatenate((x,y.T), axis=1) 

z 
```

```python
import numpy as np  
x=np.array([[1,2],[3,4]])  
y=np.array([[12,30]])  

z=np.concatenate((x,y), axis=None)  

z
```

**NumPy - Arithmetic Operations**

Functions:add(), subtract(), multiply(), and divide()


```python
x=np.array([[1,1],[1,1]])  
y=np.array([[10,10],[10,10]])
print("x:",x)
print("y:",y)
print("x+y:", np.add(x,y))
print("x-y:", np.subtract(x,y))
print("x*y:", np.multiply(x,y))
print("x/y:", np.divide(x,y))
print("Matrix multiplication of x and y:", np.matmul(x,y))
print("Dot multiplication of x and y:", np.dot(x,y))
```
